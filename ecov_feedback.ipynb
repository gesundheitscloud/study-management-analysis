```python
import os
import sys
import pandas as pd

from lib.display import Display
from lib.disk import Disk

from matplotlib import pyplot as plt, ticker
def show_chart(column_name, title, ax):    
    vc = df[column_name].value_counts()
    vc.plot(kind="bar", title=title, ax=ax, xlabel='Answers', ylabel='Users');
    Display.tag_bars(ax)
```

```python
disk = Disk('others')   
df = disk.read_by_label('ecov_usr_feedback', ',')

Display.frame(df.drop(columns=['data']), 5)
Display.frame(df[['data']], 5)
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(1,3, figsize=(18,7))

Display.header('Demographics', 'h1')

show_chart('native_language', 'Native language?', ax1)
show_chart('profession', 'What is your profession?', ax2)
show_chart('education_qualification', 'What is your education level?', ax3)
```

```python
fig, ((ax1, ax2), (ax3, ax4), (ax5, ax6)) = plt.subplots(3,2, figsize=(16,27))

Display.header('Answers', 'h1')

show_chart('how_often_willing_to_answer_questions', 'How often are you willing to answer questions for medical studies?', ax1)
show_chart('like_insights', 'Like insights', ax2)

show_chart('was_study_purpose_well_explained', 'How well did we explain the purpose of the study?', ax3)
show_chart('participate_in_other_studies', 'Would you participate in other studies?', ax4)

show_chart('contact_researcher_directly', 'Would you like to receive email updates about the research?', ax5)
show_chart('aware_of_waiting_list', 'Are you aware of the waiting list?', ax6)
```
